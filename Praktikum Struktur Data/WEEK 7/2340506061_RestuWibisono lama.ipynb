```python
class KelasPelatihan:
    def __init__(self):
        self.siswa = []

    def tambah_siswa(self, siswa):
        self.siswa.append(siswa)

    def laksanakan_wawancara(self):
        siswa_terurut = self.siswa[:]  # Salin siswa agar tidak merusak data asli
        n = len(siswa_terurut)
        for i in range(n):
            for j in range(0, n-i-1):
                if siswa_terurut[j].hasil_wawancara is None:
                    siswa_terurut[j].wawancara()
                if (siswa_terurut[j].hasil_wawancara if siswa_terurut[j].hasil_wawancara is not None else 0) < (siswa_terurut[j+1].hasil_wawancara if siswa_terurut[j+1].hasil_wawancara is not None else 0):
                    siswa_terurut[j], siswa_terurut[j+1] = siswa_terurut[j+1], siswa_terurut[j]

        siswa_gagal = []
        siswa_sempurna = []
        for siswa in siswa_terurut:
            if siswa.hasil_wawancara is None:
                siswa.wawancara()
            if siswa.hasil_wawancara <= 60:
                siswa_gagal.append(siswa)
            elif siswa.hasil_wawancara == 100:
                siswa_sempurna.append(siswa)

        if siswa_gagal:
            print("")
            print("Siswa yang gagal dalam wawancara:")
            for siswa in siswa_gagal:
                print(f"Siswa dengan NIS {siswa.nis_siswa} gagal dalam wawancara. Siswa-siswa yang gagal harus mengulang minggu depan.")

        if siswa_sempurna:
            print("")
            print("Siswa yang mendapatkan nilai sempurna dan akan mendapatkan penghargaan:")
            for siswa in siswa_sempurna:
                print(f"Siswa dengan NIS {siswa.nis_siswa}")

class Siswa:
    def __init__(self, nis_siswa):
        self.nis_siswa = nis_siswa
        self.hasil_wawancara = None

    def wawancara(self):
        if self.hasil_wawancara is None:
            self.hasil_wawancara = int(input(f"Masukkan hasil wawancara untuk siswa dengan NIS {self.nis_siswa}: "))
        print(f"Hasil yang diraih siswa NIS {self.nis_siswa}, Skor {self.hasil_wawancara}")

if __name__ == "__main__":
    kelas_pelatihan = KelasPelatihan()

    for nis_siswa in range(1, 6):
        kelas_pelatihan.tambah_siswa(Siswa(nis_siswa))

    kelas_pelatihan.laksanakan_wawancara()
```
