```python
class Node:
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None

def level_order(temp):
    if not temp:
        return
    q = []
    q.append(temp)
    while q:
        temp = q.pop(0)
        print(temp.data, end=" ")
        if temp.left:
            q.append(temp.left)
        if temp.right:
            q.append(temp.right)

def delDeep(root, d_node):
    q = []
    q.append(root)
    while(len(q)):
        temp = q.pop(0)
        if temp is d_node:
            temp = None
            return
        if temp.right:
            if temp.right is d_node:
                temp.right = None
                return
            else:
                q.append(temp.right)
        if temp.left:
            if temp.left is d_node:
                temp.left = None
                return
            else:
                q.append(temp.left)

def delet(root, key):
    if root == None:
        return None
    if root.left == None and root.right == None:
        if root.key == key:
            return None
        else:
            return root
    k_node = None
    q = []
    q.append(root)
    temp = None
    while(len(q)):
        temp = q.pop(0)
        if temp.left:
            q.append(temp.left)
        if temp.right:
            q.append(temp.right)
    if k_node:
        delDeep(root, temp)
    return root

def insert(root, data):
    q = []
    q.append(root)
    while len(q):
        temp = q.pop(0)
        if not temp.left:
            temp.left = Node(data)
            break
        else:
            q.append(temp.left)
        if not temp.right:
            temp.right = Node(data)
            break
        else:
            q.append(temp.right)

root = Node(8)
root.left = Node(3)
root.right = Node(15)
root.right.left = Node(12)
root.right.right = Node(7)

print("sebelum dihapus")
level_order(root)

root = delet(root, 12)
print("\nsetelah dihapus:")
level_order(root)

insert(root, 5)
insert(root, 6)
print("\nsetelah diinsert:")
level_order(root)
```

```python
class Node:
    def __init__(self, key, value):
        self.key = key
        self.value = value
        self.next = None

class separatechaining:
    def __init__(self, size):
        self.size = size
        self.table = [None] * size;

    def hash_function(self, key):
        return hash(key) % self.size;

    def insert(self, key, value):
        index = self.hash_function(key)
        if self.table[index] is None:
            self.table[index] = Node(key, value)
        else:
            current = self.table[index]
            while current.next:
                current = current.next
            current.next = Node(key, value)

    def search(self, key):
        index = self.hash_function(key)
        current = self.table[index]
        while current:
            if current.key == key:
                return current.value
            current = current.next
        return None

hash_table = separatechaining(10)
hash_table.insert("Paku", 5)
hash_table.insert("Palu", 10)
hash_table.insert("Gergaji", 15)
hash_table.insert("Skop", 20)
hash_table.insert("Pasah", 25)

print(hash_table.search("Skop"))
print(hash_table.search("Pasah"))
print(hash_table.search("Palu"))
```

```python
debt = {
    "Ani": {"Budi": 5000, "Doni": 10000},
    "Budi": {"Cahyo": 10000},
    "Cahyo": {"Ani": 1000, "Eko": 10000},
    "Doni": {"Cahyo": 5000},
    "Eko": {"Budi": 5000, "Doni": 500}
}

balance = {}
for people in debt:
    for creditor, amount in debt[people].items():
        if people not in balance:
            balance[people] = 0
        if creditor not in balance:
            balance[creditor] = 0
        balance[people] -= amount
        balance[creditor] += amount

debtors = [(person, amount) for person, amount in balance.items() if amount < 0]
creditors = [(person, amount) for person, amount in balance.items() if amount > 0]

debtors.sort(key=lambda x: x[1])
creditors.sort(key=lambda x: x[1], reverse=True)

transactions = []

i, j = 0, 0
while i < len(debtors) and j < len(creditors):
    debtor = debtors[i]
    creditor = creditors[j]
    if debtor[1] == 0:
        i += 1
        continue
    if creditor[1] == 0:
        j += 1
        continue
    amount = min(-debtor[1], creditor[1])
    debtor_name = debtor[0]
    creditor_name = creditor[0]
    transactions.append((debtor_name, creditor_name, amount))
    debtors[i] = (debtor_name, debtor[1] + amount)
    creditors[j] = (creditor_name, creditor[1] - amount)
    if debtors[i][1] == 0:
        i += 1
    if creditors[j][1] == 0:
        j += 1

if any(debtor[1] < 0 for debtor in debtors):
    print("Masih ada hutang yang belum terbayar")
else:
    for transaction in transactions:
        print(f"{transaction[0]} harus membayar {transaction[2]} ke {transaction[1]}")
```
