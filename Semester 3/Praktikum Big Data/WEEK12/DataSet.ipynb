```python
import pandas as pd

# Membaca file CSV dengan path yang benar
df = pd.read_csv("~/Documents/KuliahBackUP/caso_full.csv")

# Menampilkan 5 baris pertama dari data
print('Lima baris pertama dari data:\n')
print(df.head())

# Menampilkan informasi lebih lanjut tentang dataset, seperti kolom dan tipe data
print('Informasi dataset:\n')
print(df.info())

# Menampilkan statistik deskriptif dari data numerik
print('Statistik deskriptif dari data:\n')
print(df.describe())

# Menampilkan nama-nama kolom dalam dataset
print('Nama-nama kolom:\n')
print(df.columns)
```

```python
# Normalisasi menggunakan Min-Max Scaling pada kolom numerik
df['last_available_confirmed_normalized'] = (df['last_available_confirmed'] - df['last_available_confirmed'].min()) / (df['last_available_confirmed'].max() - df['last_available_confirmed'].min())
df['last_available_deaths_normalized'] = (df['last_available_deaths'] - df['last_available_deaths'].min()) / (df['last_available_deaths'].max() - df['last_available_deaths'].min())
df['new_confirmed_normalized'] = (df['new_confirmed'] - df['new_confirmed'].min()) / (df['new_confirmed'].max() - df['new_confirmed'].min())

# Menampilkan hasil normalisasi
print("Data setelah Min-Max Normalization:\n")
print(df[['last_available_confirmed', 'last_available_confirmed_normalized', 
          'last_available_deaths', 'last_available_deaths_normalized', 
          'new_confirmed', 'new_confirmed_normalized']].head())
```

```python
# Standardisasi menggunakan Z-Score pada kolom numerik
df['last_available_confirmed_standardized'] = (df['last_available_confirmed'] - df['last_available_confirmed'].mean()) / df['last_available_confirmed'].std()
df['last_available_deaths_standardized'] = (df['last_available_deaths'] - df['last_available_deaths'].mean()) / df['last_available_deaths'].std()
df['new_confirmed_standardized'] = (df['new_confirmed'] - df['new_confirmed'].mean()) / df['new_confirmed'].std()

# Menampilkan hasil standardisasi
print("Data setelah Z-Score Standardization:\n")
print(df[['last_available_confirmed', 'last_available_confirmed_standardized', 
          'last_available_deaths', 'last_available_deaths_standardized', 
          'new_confirmed', 'new_confirmed_standardized']].head())
```

```python
# Menyimpan DataFrame yang sudah dibersihkan ke file CSV
df.to_csv('~/Documents/KuliahBackUP/caso_full.csv', index=False)
```
